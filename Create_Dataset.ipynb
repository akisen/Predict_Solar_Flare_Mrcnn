```python
import sunpy.map
path = "/Volumes/TRANSCEND/dataset/HMI_REGION/201005/hmi.M_45s.20100531_220000_TAI.2.magnetogram.fits"
map = sunpy.map.Map(path)
map
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(1)
ax = plt.subplot(1,1,1)
ax.set_ylim([0, len(map.data)]); #y軸を反転
high_map = np.where(((map.data<-500) | (map.data>500)),1,0)
im = ax.imshow(high_map); #climの範囲を変えればダイナミックレンジを変更できる
#カラーバーのプロット
plt.colorbar(im)
# plt.figure(1).suptitle("")#画像の上側にタイトル
# plt.figure(1).savefig("保存するときの名前")#保存
plt.show()
```

全然値に変化がみられない→最初の分布のY軸が対数なのが原因そう
